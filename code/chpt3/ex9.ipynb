## Description

Let $X_1$, $X_2$, ... , $X_n$ be $N(0, 1)$ random variables, and let $\overline X_n = n^{-1} \sum _{i = 1}^n X_i$. Plot $\overline X_n$ versus $n$ for $n = 1, \dots, 10000$. Repeat for $X_1, X_2, ... , X_n \sim Cauchy$. Explain why there is such a difference.

```julia
using Pkg
Pkg.add("Plots")
Pkg.add("Distributions")

using Plots, Random, Distributions, Statistics
```

```julia
X_n = 1:10000 .|> n -> mean(rand(Normal(0, 1), n)) 
```

```julia
plot(1:10000, X_n)
```

```julia
Y_n = X_n = 1:10000 .|> n -> mean(rand(Cauchy(), n)) 
```

```julia
plot(1:10000, Y_n)
```

The main difference of the plots results from their variance.
